Fill NaN Age with Mean Age

```python
df = pd.read_csv(url)
```

```python
mean_age = df["Age"].mean()
mean_age
```

```python
df["Age"] = df["Age"].fillna(mean_age)
```

```python
df.isnull().sum()
```

```python
df["Family"] = df["SibSp"] + df["Parch"]
df = df.drop(columns=["SibSp","Parch"])
```

```python
df
```

### Task 1: Train-test split (80/20) from the above df

```python
from sklearn.model_selection import train_test_split
train_data,test_data = train_test_split(df,test_size=0.2)
```

### Task: Cabin should be dropped since it has so many nulls

```python
train_data= train_data.drop(columns=["Cabin"])
test_data = test_data.drop(columns=["Cabin"])
```

### Task: Extract what gender or status from the "Name" variable. Make a new feature called "title" ex: miss, mrs, mr, sir, and etc. After, drop the Name column 

```python
#Add Status column Unmarried = miss,Married = Mrs ,Unknown = Mr
train_data['Status'] = train_data['Name'].apply(lambda name: 'Unknown' if 'Mr.' in name else ('Married' if 'Mrs.' in name else 'Unmarried'))
test_data['Status'] = test_data['Name'].apply(lambda name: 'Unknown' if 'Mr.' in name else ('Married' if 'Mrs.' in name else 'Unmarried'))

#Drop columns
train_data = train_data.drop(columns=["PassengerId","Name"])
test_data = test_data.drop(columns=["PassengerId","Name"])
train_data
```

```python
test_data
```

### Task: Try to bucket age into categories (ex: baby, kid, teenager, adult, and etc)

```python
def get_title(row):
    age = row['Age']
    if age < 13:
        return 'Children'
    elif 13 <= age <= 19:
        return 'Teenager'
    elif 19< age < 60:
        return 'Adult'
    elif age>= 60:
        return 'Elderly'
    else:
        return None

train_data['Title'] = train_data.apply(get_title, axis=1)
test_data['Title'] = test_data.apply(get_title, axis=1)

```

### Task: Drop ticket

```python
train_data.drop(columns=["Ticket"],inplace=True)
test_data.drop(columns=["Ticket"],inplace=True)
```
